```python
import pandas as pd
import numpy as np

# Configuración de reproducibilidad para que los resultados sean consistentes
np.random.seed(42)
n_rows = 2000

# Tu inventario real detallado
equipos_reales = [
    {'t': 'GRUPO ELECTRÓGENO', 'm': 'TIDE POWER', 'mod': 'FC8SX-YD9'},
    {'t': 'UPS MN TS', 'm': 'POWER / SP II', 'mod': '3000 VA'},
    {'t': 'UPS TF STD ALONE', 'm': 'RIELLO / MST', 'mod': 'MST20'},
    {'t': 'UPS MN RTLIEBERT', 'm': 'GXT5', 'mod': '3000 VA'},
    {'t': 'UPS MN RTLIEBERT', 'm': 'GXT5', 'mod': '1.5KVA'},
    {'t': 'UPS TF STD ALONE', 'm': 'RIELLO / MST', 'mod': 'MST60'},
    {'t': 'UPS MN MODULAR', 'm': 'LIEBERT', 'mod': 'APS10'},
    {'t': 'UPS TF MODULAR', 'm': 'SPOWER / SP M', 'mod': '2020'},
    {'t': 'UPS MN RT', 'm': 'DELTA / GES103R', 'mod': '103R'},
    {'t': 'UPS TF STD ALONE', 'm': 'RIELLO / MST', 'mod': 'MST40'},
    {'t': 'AIRE ACONDICIONADO', 'm': 'HUAWEI', 'mod': 'Netcol5000-A013'}
]

servicios = ['Mantenimiento Preventivo', 'Mantenimiento Correctivo', 'Mantenimiento Predictivo', 'Instalación']
ubicaciones = ['Lima Centro/Sur', 'Lima Periferia/Callao', 'Provincia Cercana', 'Provincia Alejada']

# Tarifas diarias de OLC
P_SUP, P_TEC, P_APO = 250, 150, 100

registros = []

for i in range(n_rows):
    eq = np.random.choice(equipos_reales)
    ser = np.random.choice(servicios, p=[0.4, 0.3, 0.2, 0.1])
    loc = np.random.choice(ubicaciones, p=[0.4, 0.3, 0.2, 0.1])

    # Lógica de operación
    if ser == 'Instalación':
        dias = np.random.randint(3, 7)
        n_s, n_t, n_a = 1, 2, 2
    elif ser == 'Mantenimiento Correctivo':
        dias = np.random.randint(1, 4)
        n_s, n_t, n_a = 0, 2, 1
    else: # Preventivo / Predictivo
        dias = 1
        n_s, n_t, n_a = 0, 1, 1

    # Cálculo de Costo de Mano de Obra
    costo_mo = (n_s*P_SUP + n_t*P_TEC + n_a*P_APO) * dias

    # Viáticos (Factor Lima vs Provincias alejadas)
    viaticos_base = {'Lima Centro/Sur': 40, 'Lima Periferia/Callao': 60, 'Provincia Cercana': 160, 'Provincia Alejada': 320}
    costo_via = viaticos_base[loc] * (n_s + n_t + n_a) * dias

    # Factores que disparan el costo (lo que la IA debe aprender)
    es_nocturno = np.random.choice([0, 1], p=[0.8, 0.2])
    sobrecosto_nocturno = costo_mo * 0.35 if es_nocturno else 0

    requiere_repuestos = 1 if (ser == 'Mantenimiento Correctivo' and np.random.rand() > 0.4) else 0
    costo_repuestos = np.random.uniform(1200, 6000) if requiere_repuestos else 0

    # COSTO REAL TOTAL (Target para el proyecto)
    costo_real = costo_mo + sobrecosto_nocturno + costo_via + costo_repuestos + np.random.normal(0, 80)

    # PRESUPUESTO COMERCIAL (Subestimado)
    presup_comercial = (costo_mo * 0.95) + (costo_via * 0.8)

    registros.append({
        'OT_Servicio': f'OLC-2026-{i+1000}',
        'Tipo_Servicio': ser,
        'Equipo_Infraestructura': eq['t'],
        'Marca_Equipo': eq['m'],
        'Modelo_Capacidad': eq['mod'],
        'Ubicacion_Servicio': loc,
        'Es_Horario_Nocturno': es_nocturno,
        'Dias_de_Operacion': dias,
        'Uso_de_Repuestos': requiere_repuestos,
        'Presupuesto_Comercial_Soles': round(presup_comercial, 2),
        'Costo_Real_Operativo_Soles': round(max(costo_real, 120), 2)
    })

# Guardar a CSV
df_olc = pd.DataFrame(registros)
df_olc.to_csv('Dataset_OLC_Operaciones_Final.csv', index=False)
print("¡Éxito! Archivo 'Dataset_OLC_Operaciones_Final.csv' generado.")
```
